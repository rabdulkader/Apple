```python
# import the pygame module, so you can use it
import pygame
import random
import time
 
# define a main function
def main():
     
    # initialize the pygame module
    pygame.init()
     
    # create a surface on screen that has the size of 240 x 180
    screen = pygame.display.set_mode((600,600))
    
    x_r_edge=600
    x_l_edge=-1
    y_b_edge=600
    y_t_edge=-1
    
    white=[255,255,255]
    red=[255,0,0]
    black=[0,0,0]
    green=[0,255,0]
    
    # define a variable to control the main loop
    running = True
    
    x=180
    y=180
    w=60
    h=60
    
    grid_x=[]
    grid_y=[]
    for i in range(0,600,60):
        grid_x.append(i)
        
    for i in range(0,600,60):
        grid_y.append(i)
    
    mleft=False
    mup=False
    mright=False
    mdown=False
    
    p1=pygame.Rect([x,y,w,h])
    snake=[p1,p1]
    
    
    apple_present=False
    # main loop
    while running:
        print(len(snake))

        
        if  mleft==True:
            snake[0].x -=60
        if mright==True:
            snake[0].x +=60
        if mup==True:
            snake[0].y -=60 
        if mdown==True:
            snake[0].y +=60           
            
        if snake[0].x <= (x_l_edge-1):
            for i in range(len(snake)-1,0,-1):
                snake.remove(snake[i])
                
            snake[0].x=180
            snake[0].y=180
            #break
        if snake[0].y <= (y_t_edge-1):
            for i in range(len(snake)-1,0,-1):
                snake.remove(snake[i])
            
            snake[0].x=180
            snake[0].y=180
            #break
        if (snake[0].x + w) >= (x_r_edge+1):
            for i in range(len(snake)-1,0,-1):
                snake.remove(snake[i])
            
            snake[0]. x=180
            snake[0].y=180
            #break
        if (snake[0].y + h) >= (y_b_edge+1):
            for i in range(len(snake)-1,0,-1):
                snake.remove(snake[i])
            
            snake[0].x=180
            snake[0].y=180     
            #break
            

        

                         
        for event in pygame.event.get():
            # only do something if the event is of type QUIT
            
            if event.type == pygame.KEYDOWN:
                if event.key == pygame.K_LEFT:
                    #x -=10
                    mleft=True
                    mup=False
                    mright=False
                    mdown=False
                    print('key left')
                if event.key == pygame.K_RIGHT:
                    #x +=10
                    mright=True
                    mup=False
                    mleft=False
                    mdown=False
                    print('key Right')
                if event.key == pygame.K_UP:
                    #y -=10
                    mup=True
                    mright=False
                    mleft=False
                    mdown=False
                    print('key Up')
                if event.key == pygame.K_DOWN:
                   #y +=10
                    mdown=True
                    mleft=False
                    mright=False
                    mup=False
                    print('key Down')
                if event.key == pygame.K_0:
                    i=len(snake)-1
                    part=pygame.Rect([snake[i].x,snake[i].y,w,h])
                    snake.append(part)
                    

            
            if event.type == pygame.QUIT:
                # change the value to False, to exit the main loop
                running = False
                
                #if event.key == pygame.K_1:
        if apple_present ==False:
            for i in range(0,len(snake)):
                apple_x=random.choice(grid_x)
                if apple_x==snake[i].x:
                    apple_x=random.choice(grid_x)

                apple_y=random.choice(grid_y)
                if apple_y==snake[i].y:
                    apple_y=random.choice(grid_y)

            apple=pygame.Rect([apple_x,apple_y,w,h])
            apple_present=True
            
        screen.fill(white)
        
        if snake[0].x == apple.x and snake[0].y == apple.y:
            i=len(snake)-1
            part=pygame.Rect([snake[i].x,snake[i].y,w,h])
            snake.append(part)
            apple_present=False
        

        
        
        for i in range(len(snake)-1,-1,-1):
            if i==0:
                pygame.draw.rect(screen,red,snake[i])
                break
            snake[i].x=snake[i-1].x
            snake[i].y=snake[i-1].y
            pygame.draw.rect(screen,black,snake[i])
        
        pygame.draw.rect(screen,green,apple)
    
        pygame.display.update()

        time.sleep(0.15)
        

    
main()
```
