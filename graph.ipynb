```python
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize']=[8,6]

def draw(scores):
    plt.xlim((0,1300))
    plt.ylim((0,1000))
    for j in range(len(scores)):
        a=scores[j]
        #print(a)
        for i in range(len(a)):
            plt.scatter(a[i][0],a[i][1],c='red',s=20)
    #plt.scatter(virus[0], virus[1], c ='red', s = 60)


scores=[None]*1219
for j in range(1,1219):
    
    file=open(r'scores_7\gen_'+str(j)+'.txt','r')
    score_str=file.read()
    file.close()

    refined_score=score_str.split()
    score_list=[None]*len(refined_score)
    
    for i in range(len(refined_score)):
        if i == 0:
            score_list[i]=j,float(refined_score[i].split('[')[1].split(',')[0])*-0.001

        elif i == len(refined_score) - 1:
            score_list[i]=j,float(refined_score[i].split(']')[0])*-0.001

        else:
            score_list[i]=j,float(refined_score[i].split(',')[0])*-0.001
    
    scores[j]=score_list
    
draw(scores[1:])
#scores
```
